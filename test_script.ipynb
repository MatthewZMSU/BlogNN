```python
import re

pattern = re.compile(r'[Нн]а самом деле|[Вв] общем|\w+-\w+|\w+')
```

```python
uncleaned = re.findall(pattern, text_1)
```

```python
def cleaning(word: str) -> bool:
    if not word:
        return False
    if re.match(r'.*\d+.*', word):
        return False
    return True

cleaned = filter(cleaning, uncleaned)
```

```python
normalized = map(lambda word: word.lower(), cleaned)
```

```python
from nltk.stem.snowball import RussianStemmer

stemmer = RussianStemmer()

stemmed = map(lambda word: stemmer.stem(word), normalized)
```

```python
words = list(stemmed)
```

```python
from collections import Counter

document_words = Counter(words)
```

```python
corpus_words = Counter(set(words))
```

```python
corpus_words
```

```python
key_words = list(corpus_words.keys())
```

```python
number_of_documents = 2
```

```python
import numpy as np

features = np.zeros(shape=(len(key_words), ), dtype=float)
for ind, word in enumerate(key_words):
    tf = document_words[word] / document_words.total()
    idf = np.log(number_of_documents / corpus_words[word])
    features[ind] = tf * idf
```
